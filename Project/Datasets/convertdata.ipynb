```python
import pandas as pd
```

```python
data = pd.read_csv('unprepareddata.csv')
prepareddata = pd.DataFrame()
# columns of this dataframe are name , year , value , lastValue
# name is the name of the country
# year is the year of the data
# value is the value of the data
# lastValue is the value of the data of the previous year
data['Area'] = data['Area'].astype('float')
for i in range(0, len(data)):
    data[str(j)] = data[str(j)].astype('float')
    # for 1990 , lastValue is the current value
    prepareddata = prepareddata.append({'name': data['Name'][i], 'year': '1990', 'value': 100*data['1990'][i]/int(data['Area'][i]), 'lastValue': 100*data['1990'][i]/int(data['Area'][i])}, ignore_index=True)
    print(type(data['Area'][i]),data['Area'][i])
    for j in range(1991, 2021):
        prepareddata = prepareddata.append({'name': data['Name'][i], 'year': j, 'value':100* data[str(j)][i]/int(data['Area'][i]), 'lastValue': 100*data[str(j-1)][i]/int(data['Area'][i])}, ignore_index=True)

prepareddata.to_csv('racebardata.csv', index=False)
```
